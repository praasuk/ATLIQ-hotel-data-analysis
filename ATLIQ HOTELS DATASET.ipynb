```python
import pandas as pd

# Ensure booking_date is datetime
df5['booking_date'] = pd.to_datetime(df5['booking_date'])

# Extract month abbreviation in lowercase (jan, feb, etc.)
df5['month_abbr'] = df5['booking_date'].dt.strftime('%b').str.lower()

# Calculate week of the month (1 to 5)
df5['week_of_month'] = df5['booking_date'].apply(lambda d: (d.day - 1) // 7 + 1)

# Get weekday number (Monday=1, Sunday=7)
df5['weekday'] = df5['booking_date'].dt.weekday + 1

# Create month_week_label only if weekday is between 1 and 4 (Mon-Thu)
df5['month_week_label'] = df5.apply(
    lambda row: f"{row['month_abbr']} w{row['week_of_month']}" if row['weekday'] in [1, 2, 3, 4] else None,
    axis=1
)

# Drop helper columns if you want
df5.drop(['month_abbr', 'week_of_month', 'weekday'], axis=1, inplace=True)
```

Customer Segmentation by Booking Patterns Using booking frequency and average lead time, segment customers into groups (e.g., early planners, last-minute bookers, frequent vs. occasional). Analyze how these segments differ in revenue and ratings.



```python
df5.head(2)
```

```python
# Step 1: Group by property_id and booking_platform
segment_df = df5.groupby(['property_id', 'booking_platform']).agg({
    'booking_id': 'count',               # total bookings
    'lead_time': 'mean',                 # avg lead time
    'revenue_generated': 'sum',         # total revenue
    'ratings_given': 'mean'             # avg rating
}).reset_index()

# Step 2: Rename columns
segment_df.rename(columns={
    'booking_id': 'total_bookings',
    'lead_time': 'avg_lead_time',
    'revenue_generated': 'total_revenue',
    'ratings_given': 'avg_rating'
}, inplace=True)

# Step 3: Calculate booking frequency
total_groups = segment_df.shape[0]
segment_df['booking_frequency'] = round((segment_df['total_bookings'] / total_groups) * 100, 2)

# Step 4: Create segments
segment_df['planner_type'] = pd.cut(segment_df['avg_lead_time'],
                                    bins=[0, 7, 21, 60, 365],
                                    labels=['Last-minute', 'Short-term', 'Medium-term', 'Early Planner'])

segment_df['booking_type'] = pd.cut(segment_df['booking_frequency'],
                                    bins=[0, 5, 15, 100],
                                    labels=['Occasional', 'Regular', 'Frequent'])

```

```python
df5.columns
```

```python
segment_df
```

```python
# Revenue and Ratings by Planner Type
planner_analysis = segment_df.groupby('planner_type').agg({
    'total_revenue': 'mean',
    'avg_rating': 'mean'
}).reset_index()

# Revenue and Ratings by Booking Type
booking_analysis = segment_df.groupby('booking_type').agg({
    'total_revenue': 'mean',
    'avg_rating': 'mean'
}).reset_index()

# Optionally round the values
planner_analysis = planner_analysis.round(2)
booking_analysis = booking_analysis.round(2)

print("📊 Revenue & Ratings by Planner Type:")
print(planner_analysis)

print("\n📊 Revenue & Ratings by Booking Type:")
print(booking_analysis)
```
